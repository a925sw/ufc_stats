```python
import pandas as pd
import numpy as np
import os
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import resample
import pickle 
pd.set_option('display.max_columns', None)
```

```python
big_df = pd.read_csv("../data/big_df.csv")
big_df
```

```python
train_test_df = big_df[big_df["winner"] != "tbd"]
train_test_df
```

```python
number = LabelEncoder()
# data['red_name'] = number.fit_transform(data['red_name'].astype('str'))
# data['blue_name'] = number.fit_transform(data['blue_name'].astype('str'))
train_test_df['red_stance'] = number.fit_transform(train_test_df['red_stance'].astype('str'))
train_test_df['blue_stance'] = number.fit_transform(train_test_df['blue_stance'].astype('str'))
train_test_df['weight_class'] = number.fit_transform(train_test_df['weight_class'].astype('str'))
train_test_df
```

```python
data = train_test_df.drop(columns=["lev1", "winner", \
                                   "red_name", "blue_name"
                                  ])
target = train_test_df["winner"]
feature_names = data.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
sorted(zip(clf.feature_importances_, data), reverse=True)
```

```python
rf = RandomForestClassifier(n_estimators=105)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
preds = rf.predict(data)
```

```python
pred_df = pd.DataFrame({"actual": train_test_df["winner"],
                       "pred" : preds})
pred_df.to_csv("../data/pred_df.csv")
pred_df
```

```python
train_test_df["pred"] = preds
train_test_df.sample(100)
```

```python
pred_df = train_test_df.drop(columns="lev1")
pred_df.to_csv("../data/rf_preds.csv")
pred_df
```

```python
sorted(zip(rf.feature_importances_, data), reverse=True)
```

```python
feature_importance_df = pd.DataFrame(sorted(zip(feature_names, rf.feature_importances_), reverse=True), \
                                     columns = ['feature_names', 'feature_importance'])
feature_importance_df = feature_importance_df.sort_values(by='feature_importance', ascending=False).reset_index(drop=True)
# feature_importance_df.to_csv('../data/feature_importance_random_forest_df.csv')
feature_importance_df
```
