```python
# Load in the previously cleaned up and reformatted fights data
clean_df = pd.read_csv("../data/cleaned_fight_df.csv")
clean_df = clean_df.drop(columns="Unnamed: 0")
clean_df
```

```python
# Load in event data to merge in with fights
event_df = pd.read_csv("../data/rolling_event_data.csv")
event_df = event_df.drop(columns="Unnamed: 0")
event_df = event_df.rename(columns={"url": "event_url"})
event_df
```

```python
# add future event to all events
event_df = pd.concat([future_event_df, event_df])
event_df
```

```python
# Merging events with fights
clean_df = clean_df.merge(event_df, on="event")
clean_df
```

```python
# Dropping all fights that occured before unified rules were implamented 
clean_df["date"] = pd.to_datetime(clean_df["date"])
clean_df = clean_df[clean_df["date"] >= "November 17, 2000"]
clean_df
```

```python
clean_df["method"].value_counts()
```

```python
# streamlining labels of fight outcomes
clean_df["method"] = clean_df["method"].replace({"TKO - Doctor's Stoppage" : "KO/TKO",
                                                "Could Not Continue" : "KO/TKO",
                                                "Decision - Majority" : "MD",
                                                "Decision - Split" : "SD",
                                                "Decision - Unanimous": "UD",
                                                "Other": "n/a",
                                                "DQ" : "n/a",
                                                "Overturned": "n/a"
                                                })
clean_df["method"].value_counts()
```

```python
# removing fights without useful outcome
clean_df = clean_df[clean_df["method"] != "n/a"]
clean_df.sample(10)
```

```python
# creating columns of win/loss outcome that can be summed later on
clean_df["red_win"] = clean_df["winner"].apply(lambda x: 1 if x == "red" else (0 if x == "tbd" else 0))
clean_df["blue_win"] = clean_df["winner"].apply(lambda x: 1 if x == "blue" else (0 if x == "tbd" else 0))
clean_df["red_loss"] = clean_df["winner"].apply(lambda x: 1 if x == "blue" else (0 if x == "tbd" else 0))
clean_df["blue_loss"] = clean_df["winner"].apply(lambda x: 1 if x == "red" else (0 if x == "tbd" else 0))

clean_df.head(20)
```

```python
# removing draws and no contests 
clean_df = clean_df[clean_df["winner"] != "Draw"] # it's arguable draws are useful datapoints, but I choose to remove them
clean_df = clean_df[clean_df["winner"] != "NC"] # no contests however are not useful datapoints
clean_df["winner"].value_counts()
```

```python
clean_df
```

```python
# creating lists to hold summable fight outcomes 
red_kos = []
blue_kos = []
red_subs = []
blue_subs = []
red_decs = []
blue_decs = []

# Looping through fights to determine winner and method to use .cumsum() method on later
for row in clean_df.itertuples():
    if row.red_win == 0 & row.blue_win == 0:
        red_ko = 0
        red_sub = 0
        red_dec = 0
        blue_ko = 0
        blue_sub = 0
        blue_dec = 0
        
        red_kos.append(red_ko)
        red_subs.append(red_sub)
        red_decs.append(red_dec)
        blue_kos.append(blue_ko)
        blue_subs.append(blue_sub)
        blue_decs.append(blue_dec)
    elif row.red_win == 1:
        method = row.method
        if method == "KO/TKO":
            red_ko = 1
            red_sub = 0
            red_dec = 0
        elif method == "Submission":
            red_ko = 0
            red_sub = 1
            red_dec = 0
        else:
            red_ko = 0
            red_sub = 0
            red_dec = 1
        red_kos.append(red_ko)
        red_subs.append(red_sub)
        red_decs.append(red_dec)
        blue_kos.append(0)
        blue_subs.append(0)
        blue_decs.append(0)
    elif row.blue_win == 1:
        method = row.method
        if method == "KO/TKO":
            blue_ko = 1
            blue_sub = 0
            blue_dec = 0
        elif method == "Submission":
            blue_ko = 0
            blue_sub = 1
            blue_dec = 0
        else:
            blue_ko = 0
            blue_sub = 0
            blue_dec = 1
        blue_kos.append(blue_ko)
        blue_subs.append(blue_sub)
        blue_decs.append(blue_dec)
        red_kos.append(0)
        red_subs.append(0)
        red_decs.append(0) 
    else:
        print("oops")
```

```python
# adding fight outcome columns
clean_df["red_KO/TKOWin"] = red_kos
clean_df["red_subWin"] = red_subs
clean_df["red_decWin"] = red_decs
clean_df["blue_KO/TKOWin"] = blue_kos
clean_df["blue_subWin"] = blue_subs
clean_df["blue_decWin"] = blue_decs
clean_df
```

```python
# adding the flip side of method lost for losing fighter
clean_df["red_KOLoss"] = clean_df["blue_KO/TKOWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df["blue_KOLoss"] = clean_df["red_KO/TKOWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df["red_SubLoss"] = clean_df["blue_subWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df["blue_SubLoss"] = clean_df["red_subWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df["red_DecLoss"] = clean_df["blue_decWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df["blue_DecLoss"] = clean_df["red_decWin"].apply(lambda x: 1 if x == 1 else 0)
clean_df
```

```python
# Now that each fight is stacked with nested levels for red and blue fighters, it is possible to sum up the total
# stats of each fighter by name REGARDLESS of if the were in the red or blue corner. Without this stacking, trying
# to sum up the total for any given stat (e.g takedowns) would return two numbers, one for their total when 
# fighting out of the red corner, and another for them fighting out of the blue corner. This turned into a bit of
# a conceptual challenge for me, so shout out stackoverflow user jezrael for helping me with this.

# utilizing groupby["name"], i was able to get the sum of relevant stats for each fighter through their UFC careers
# and subtract the data of the fight in that row to get a cumulative sum COMING INTO THE FIGHT, so the future model
# would only have information about the fighters that was available BEFORE the fight took place
df1["cumsum_tot_stks_lnd"] = (df1.iloc[::-1].groupby(["name"])["TotStksLnd"].cumsum()) - df1["TotStksLnd"]
df1["cumsum_tot_stks_att"] = (df1.iloc[::-1].groupby(["name"])["TotStksAtt"].cumsum()) - df1["TotStksAtt"]
df1["cumsum_sig_stks_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStksLnd"].cumsum()) - df1["SigStksLnd"]
df1["cumsum_sig_stks_att"] = (df1.iloc[::-1].groupby(["name"])["SigStksAtt"].cumsum()) - df1["SigStksAtt"]
df1["cumsum_TDs_lnd"] = (df1.iloc[::-1].groupby(["name"])["TDsLnd"].cumsum()) - df1["TDsLnd"]
df1["cumsum_TDs_att"] = (df1.iloc[::-1].groupby(["name"])["TDsAtt"].cumsum()) - df1["TDsAtt"]
df1["cumsum_sub_att"] = (df1.iloc[::-1].groupby(["name"])["SubAtts"].cumsum()) - df1["SubAtts"]
df1["cumsum_ctrl_sec"] = (df1.iloc[::-1].groupby(["name"])["CtrlSec"].cumsum()) - df1["CtrlSec"]
df1["cumsum_KDs"] = (df1.iloc[::-1].groupby(["name"])["KDs"].cumsum()) - df1["KDs"]
df1["cumsum_revs"] = (df1.iloc[::-1].groupby(["name"])["Reversals"].cumsum()) - df1["Reversals"]
df1["cumsum_sig_body_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkBodyLnd"].cumsum()) - df1["SigStkBodyLnd"]
df1["cumsum_sig_body_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkBodyAtt"].cumsum()) - df1["SigStkBodyAtt"]
df1["cumsum_sig_head_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkHeadLnd"].cumsum()) - df1["SigStkHeadLnd"]
df1["cumsum_sig_head_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkHeadAtt"].cumsum()) - df1["SigStkHeadAtt"]
df1["cumsum_sig_leg_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkLegLnd"].cumsum()) - df1["SigStkLegLnd"]
df1["cumsum_sig_leg_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkLegAtt"].cumsum()) - df1["SigStkLegAtt"]
df1["cumsum_sig_dist_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkDistLnd"].cumsum()) - df1["SigStkDistLnd"]
df1["cumsum_sig_dist_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkDistAtt"].cumsum()) - df1["SigStkDistAtt"]
df1["cumsum_sig_clch_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkClchLnd"].cumsum()) - df1["SigStkClchLnd"]
df1["cumsum_sig_clch_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkClchAtt"].cumsum()) - df1["SigStkClchAtt"]
df1["cumsum_sig_gnd_lnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkGndLnd"].cumsum()) - df1["SigStkGndLnd"]
df1["cumsum_sig_gnd_att"] = (df1.iloc[::-1].groupby(["name"])["SigStkGndAtt"].cumsum()) - df1["SigStkGndAtt"]

df1["cumsum_tot_stks_abs"] = (df1.iloc[::-1].groupby(["name"])["TotStkAbs"].cumsum()) - df1["TotStkAbs"]
df1["cumsum_tot_stks_seen"] = (df1.iloc[::-1].groupby(["name"])["TotStkSeen"].cumsum()) - df1["TotStkSeen"]
df1["cumsum_sig_stks_abs"] = (df1.iloc[::-1].groupby(["name"])["SigStkAbs"].cumsum()) - df1["SigStkAbs"]
df1["cumsum_sig_stks_seen"] = (df1.iloc[::-1].groupby(["name"])["SigStkSeen"].cumsum()) - df1["SigStkSeen"]

df1["cumsum_TDs_abs"] = (df1.iloc[::-1].groupby(["name"])["TakenDown"].cumsum()) - df1["TakenDown"]
df1["cumsum_TDs_seen"] = (df1.iloc[::-1].groupby(["name"])["TDsSeen"].cumsum()) - df1["TDsSeen"]

df1["cumsum_sub_esc"] = (df1.iloc[::-1].groupby(["name"])["SubEsc"].cumsum()) - df1["SubEsc"]
df1["cumsum_ctrld_sec"] = (df1.iloc[::-1].groupby(["name"])["CtrldSec"].cumsum()) - df1["CtrldSec"]

df1["cumsum_KedD"] = (df1.iloc[::-1].groupby(["name"])["KedD"].cumsum()) - df1["KedD"]
df1["cumsum_reversed"] = (df1.iloc[::-1].groupby(["name"])["reversed"].cumsum()) - df1["reversed"]

df1["cumsum_sig_body_abs"] = (df1.iloc[::-1].groupby(["name"])["SigBodyStkAbs"].cumsum()) - df1["SigBodyStkAbs"]
df1["cumsum_sig_body_seen"] = (df1.iloc[::-1].groupby(["name"])["SigBodyStkSeen"].cumsum()) - df1["SigBodyStkSeen"]
df1["cumsum_sig_head_abs"] = (df1.iloc[::-1].groupby(["name"])["SigHeadStkAbs"].cumsum()) - df1["SigHeadStkAbs"]
df1["cumsum_sig_head_seen"] = (df1.iloc[::-1].groupby(["name"])["SigHeadStkSeen"].cumsum()) - df1["SigHeadStkSeen"]
df1["cumsum_sig_leg_abs"] = (df1.iloc[::-1].groupby(["name"])["SigLegStkAbs"].cumsum()) - df1["SigLegStkAbs"]
df1["cumsum_sig_leg_seen"] = (df1.iloc[::-1].groupby(["name"])["SigLegStkSeen"].cumsum()) - df1["SigLegStkSeen"]
df1["cumsum_sig_dist_abs"] = (df1.iloc[::-1].groupby(["name"])["SigDistStkAbs"].cumsum()) - df1["SigDistStkAbs"]
df1["cumsum_sig_dist_seen"] = (df1.iloc[::-1].groupby(["name"])["SigDistStkSeen"].cumsum()) - df1["SigDistStkSeen"]
df1["cumsum_sig_clch_abs"] = (df1.iloc[::-1].groupby(["name"])["SigClchStkAbs"].cumsum()) - df1["SigClchStkAbs"]
df1["cumsum_sig_clch_seen"] = (df1.iloc[::-1].groupby(["name"])["SigClchStkSeen"].cumsum()) - df1["SigClchStkSeen"]
df1["cumsum_sig_gnd_abs"] = (df1.iloc[::-1].groupby(["name"])["SigGndStkAbs"].cumsum()) - df1["SigGndStkAbs"]
df1["cumsum_sig_gnd_seen"] = (df1.iloc[::-1].groupby(["name"])["SigGndStkSeen"].cumsum()) - df1["SigGndStkSeen"]
df1["cumsum_fight_sec"] = (df1.iloc[::-1].groupby(["name"])["FightSec"].cumsum()) - df1["FightSec"]

df1["cumsum_decWin"] = (df1.iloc[::-1].groupby(["name"])["decWin"].cumsum()) - df1["decWin"]
df1["cumsum_subWin"] = (df1.iloc[::-1].groupby(["name"])["subWin"].cumsum()) - df1["subWin"]
df1["cumsum_KOWin"] = (df1.iloc[::-1].groupby(["name"])["KO/TKOWin"].cumsum()) - df1["KO/TKOWin"]
df1["cumsum_Decloss"] = (df1.iloc[::-1].groupby(["name"])["DecLoss"].cumsum()) - df1["DecLoss"]
df1["cumsum_subLoss"] = (df1.iloc[::-1].groupby(["name"])["SubLoss"].cumsum()) - df1["SubLoss"]
df1["cumsum_KOLoss"] = (df1.iloc[::-1].groupby(["name"])["KOLoss"].cumsum()) - df1["KOLoss"]
df1["wins"] = (df1.iloc[::-1].groupby(["name"])["win"].cumsum()) - df1["win"]
df1["losses"] = (df1.iloc[::-1].groupby(["name"])["loss"].cumsum()) - df1["loss"]

df1["titleFights"] = (df1.iloc[::-1].groupby(["name"])["titleFight"].cumsum()) - df1["titleFight"]

df1
```
