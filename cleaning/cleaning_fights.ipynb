```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Load in scraped fights data
fight_df = pd.read_csv("../data/rolling_fights_data.csv")
fight_df = fight_df.drop(columns="Unnamed: 0")
fight_df
```

```python
# Load upcoming fights data
future_df = pd.read_csv("../data/future_fights_data.csv")
future_df = future_df.drop(columns="Unnamed: 0")
future_df
```

```python
# concat the future fights on top of past fights
fight_df = pd.concat([future_df, fight_df])
fight_df
```

```python
fight_df = fight_df.drop(columns=["url"])
fight_df.head(20)
```

```python
# Begin cleaning by stripping text out of round and time columns
fight_df["round"] = fight_df["round"].str.strip("ROUND: ")
fight_df["Time"] = fight_df["Time"].str.strip("TIME: ")
fight_df
```

```python
# Create a title bout column based off weight class text 
fight_df["title_bout"] = fight_df["weight_class"].str.contains("TITLE")
fight_df[fight_df["title_bout"] == True]
```

```python
weights = fight_df["weight_class"].value_counts()
weight_df = pd.DataFrame(weights)
weight_df
```

```python
# converting all weightclasses to set of recognizable modern divisions 
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Strawweight" if "WOMEN'S STRAWWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Flyweight" if "WOMEN'S FLY" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Bantamweight" if "WOMEN'S BAN" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Featherweight" if "WOMEN'S FEAT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Flyweight" if "FLYWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Bantamweight" if "BANTAMWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Featherweight" if "FEATHERWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Lightweight" if "LIGHTWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Welterweight" if "WELTERWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Middleweight" if "MIDDLEWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Lightheavyweight" if "LIGHT HEAVY" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Heavyweight" if "HEAVYWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Openweight" if "OPEN" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Catchweight" if "CATCH" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Tournament" if "TOURNAMENT" in x else x)
fight_df
```

```python
weights = fight_df["weight_class"].value_counts()
weight_df = pd.DataFrame(weights)
weight_df
```

```python
# splitting significant/total strikes and takedowns into landed and attempted columns
fight_df[["red_total_stks_landed","red_total_stks_attempted"]] = fight_df.red_Total_strikes.str.split(" of ",expand=True)
fight_df[["red_sig_stks_landed","red_sig_stks_attempted"]] = fight_df.red_Significant_strikes.str.split(" of ",expand=True)
fight_df[["red_TDs_landed","red_TDs_attempted"]] = fight_df.red_TDs.str.split(" of ",expand=True)
fight_df[["blue_total_stks_landed","blue_total_stks_attempted"]] = fight_df.blue_Total_strikes.str.split(" of ",expand=True)
fight_df[["blue_sig_stks_landed","blue_sig_stks_attempted"]] = fight_df.blue_Significant_strikes.str.split(" of ",expand=True)
fight_df[["blue_TDs_landed","blue_TDs_attempted"]] = fight_df.blue_TDs.str.split(" of ",expand=True)
fight_df
```

```python
fight_df = fight_df.fillna(0)
fight_df
```

```python
# Splitting strikes into landed and attempted by target and range
fight_df[["red_sig_stks_head_landed","red_sig_stks_head_attempted"]] = fight_df.red_Sig_Str_head.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_body_landed","red_sig_stks_body_attempted"]] = fight_df.red_Sig_Str_body.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_leg_landed","red_sig_stks_leg_attempted"]] = fight_df.red_Sig_Str_leg.str.split(" of ",expand=True).astype(int)


fight_df[["blue_sig_stks_head_landed","blue_sig_stks_head_attempted"]] = fight_df.red_Sig_Str_head.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_body_landed","blue_sig_stks_body_attempted"]] = fight_df.red_Sig_Str_body.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_leg_landed","blue_sig_stks_leg_attempted"]] = fight_df.red_Sig_Str_leg.str.split(" of ",expand=True).astype(int)



fight_df[["red_sig_stks_distance_landed","red_sig_stks_distance_attempted"]] = fight_df.red_Sig_Str_distance.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_clinch_landed","red_sig_stks_clinch_attempted"]] = fight_df.red_Sig_Str_clinch.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_ground_landed","red_sig_stks_ground_attempted"]] = fight_df.red_Sig_Str_ground.str.split(" of ",expand=True).astype(int)


fight_df[["blue_sig_stks_distance_landed","blue_sig_stks_distance_attempted"]] = fight_df.red_Sig_Str_distance.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_clinch_landed","blue_sig_stks_clinch_attempted"]] = fight_df.red_Sig_Str_clinch.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_ground_landed","blue_sig_stks_ground_attempted"]] = fight_df.red_Sig_Str_ground.str.split(" of ",expand=True).astype(int)


fight_df.info()
```

```python
fight_df = fight_df.fillna("0")
fight_df
```

```python
# Converting a bunch of columns to string for easier use later
fight_df["weight_class"] = pd.Series(fight_df["weight_class"], dtype="string")
fight_df["winner"] = pd.Series(fight_df["winner"], dtype="string")
fight_df["event"] = pd.Series(fight_df["event"], dtype="string")
fight_df["method"] = pd.Series(fight_df["method"], dtype="string")
fight_df["round"] = pd.Series(fight_df["round"], dtype="string")
fight_df["red_Significant_strikes"] = pd.Series(fight_df["red_Significant_strikes"], dtype="string")
fight_df["blue_Significant_strikes"] = pd.Series(fight_df["blue_Significant_strikes"], dtype="string")
fight_df["red_Total_strikes"] = pd.Series(fight_df["red_Total_strikes"], dtype="string")
fight_df["blue_Total_strikes"] = pd.Series(fight_df["blue_Total_strikes"], dtype="string")
fight_df["red_Significant_strikes"] = pd.Series(fight_df["red_Significant_strikes"], dtype="string")
fight_df["red_Sig_Str_head"] = pd.Series(fight_df["red_Sig_Str_head"], dtype="string")
fight_df["red_Sig_Str_body"] = pd.Series(fight_df["red_Sig_Str_body"], dtype="string")
fight_df["red_Sig_Str_leg"] = pd.Series(fight_df["red_Sig_Str_leg"], dtype="string")
fight_df["blue_Sig_Str_head"] = pd.Series(fight_df["blue_Sig_Str_head"], dtype="string")
fight_df["blue_Sig_Str_body"] = pd.Series(fight_df["blue_Sig_Str_body"], dtype="string")
fight_df["blue_Sig_Str_leg"] = pd.Series(fight_df["blue_Sig_Str_leg"], dtype="string")
fight_df["red_Sig_Str_distance"] = pd.Series(fight_df["red_Sig_Str_distance"], dtype="string")
fight_df["red_Sig_Str_clinch"] = pd.Series(fight_df["red_Sig_Str_clinch"], dtype="string")
fight_df["red_Sig_Str_ground"] = pd.Series(fight_df["red_Sig_Str_ground"], dtype="string")
fight_df["blue_Sig_Str_distance"] = pd.Series(fight_df["blue_Sig_Str_distance"], dtype="string")
fight_df["blue_Sig_Str_clinch"] = pd.Series(fight_df["blue_Sig_Str_clinch"], dtype="string")
fight_df["blue_Sig_Str_ground"] = pd.Series(fight_df["blue_Sig_Str_ground"], dtype="string")
fight_df.info()
```
