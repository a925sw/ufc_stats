```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Load in scraped fights data
fight_df = pd.read_csv("../data/rolling_fights_data.csv")
fight_df = fight_df.drop(columns="Unnamed: 0")
fight_df
```

```python
# Load upcoming fights data
future_df = pd.read_csv("../data/future_fights_data.csv")
future_df = future_df.drop(columns="Unnamed: 0")
future_df
```

```python
# concat the future fights on top of past fights to preform same opperations together
fight_df = pd.concat([future_df, fight_df])
fight_df
```

```python
# Dropping the url 
fight_df = fight_df.drop(columns=["url"])
fight_df.head(20)
```

```python
# Begin cleaning by stripping text out of round and time columns
fight_df["round"] = fight_df["round"].str.strip("ROUND: ")
fight_df["Time"] = fight_df["Time"].str.strip("TIME: ")
fight_df
```

```python
# Create a title bout column based off weight class text 
fight_df["title_bout"] = fight_df["weight_class"].str.contains("TITLE")
fight_df[fight_df["title_bout"] == True]
```

```python
# observing all the different "weight classes" in dataset 
weights = fight_df["weight_class"].value_counts()
weight_df = pd.DataFrame(weights)
weight_df
```

```python
# converting all weightclasses to set of recognizable divisions 
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Strawweight" if "WOMEN'S STRAWWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Flyweight" if "WOMEN'S FLY" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Bantamweight" if "WOMEN'S BAN" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Women's Featherweight" if "WOMEN'S FEAT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Flyweight" if "FLYWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Bantamweight" if "BANTAMWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Featherweight" if "FEATHERWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Lightweight" if "LIGHTWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Welterweight" if "WELTERWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Middleweight" if "MIDDLEWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Lightheavyweight" if "LIGHT HEAVY" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Men's Heavyweight" if "HEAVYWEIGHT" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Openweight" if "OPEN" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Catchweight" if "CATCH" in x else x)
fight_df["weight_class"] = fight_df["weight_class"].apply(lambda x: "Tournament" if "TOURNAMENT" in x else x)
fight_df
```

```python
weights = fight_df["weight_class"].value_counts()
weight_df = pd.DataFrame(weights)
weight_df
```

```python
# splitting significant/total strikes and takedowns into landed and attempted columns
fight_df[["red_total_stks_landed","red_total_stks_attempted"]] = fight_df.red_Total_strikes.str.split(" of ",expand=True)
fight_df[["red_sig_stks_landed","red_sig_stks_attempted"]] = fight_df.red_Significant_strikes.str.split(" of ",expand=True)
fight_df[["red_TDs_landed","red_TDs_attempted"]] = fight_df.red_TDs.str.split(" of ",expand=True)

fight_df[["blue_total_stks_landed","blue_total_stks_attempted"]] = fight_df.blue_Total_strikes.str.split(" of ",expand=True)
fight_df[["blue_sig_stks_landed","blue_sig_stks_attempted"]] = fight_df.blue_Significant_strikes.str.split(" of ",expand=True)
fight_df[["blue_TDs_landed","blue_TDs_attempted"]] = fight_df.blue_TDs.str.split(" of ",expand=True)
fight_df
```

```python
fight_df = fight_df.fillna(0)
fight_df
```

```python
# Splitting strikes into landed and attempted by target and range
fight_df[["red_sig_stks_head_landed","red_sig_stks_head_attempted"]] = fight_df.red_Sig_Str_head.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_body_landed","red_sig_stks_body_attempted"]] = fight_df.red_Sig_Str_body.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_leg_landed","red_sig_stks_leg_attempted"]] = fight_df.red_Sig_Str_leg.str.split(" of ",expand=True).astype(int)


fight_df[["blue_sig_stks_head_landed","blue_sig_stks_head_attempted"]] = fight_df.blue_Sig_Str_head.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_body_landed","blue_sig_stks_body_attempted"]] = fight_df.blue_Sig_Str_body.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_leg_landed","blue_sig_stks_leg_attempted"]] = fight_df.blue_Sig_Str_leg.str.split(" of ",expand=True).astype(int)



fight_df[["red_sig_stks_distance_landed","red_sig_stks_distance_attempted"]] = fight_df.red_Sig_Str_distance.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_clinch_landed","red_sig_stks_clinch_attempted"]] = fight_df.red_Sig_Str_clinch.str.split(" of ",expand=True).astype(int)
fight_df[["red_sig_stks_ground_landed","red_sig_stks_ground_attempted"]] = fight_df.red_Sig_Str_ground.str.split(" of ",expand=True).astype(int)


fight_df[["blue_sig_stks_distance_landed","blue_sig_stks_distance_attempted"]] = fight_df.blue_Sig_Str_distance.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_clinch_landed","blue_sig_stks_clinch_attempted"]] = fight_df.blue_Sig_Str_clinch.str.split(" of ",expand=True).astype(int)
fight_df[["blue_sig_stks_ground_landed","blue_sig_stks_ground_attempted"]] = fight_df.blue_Sig_Str_ground.str.split(" of ",expand=True).astype(int)


fight_df.info()
```

```python
fight_df = fight_df.fillna("0")
fight_df
```

```python
# Converting a bunch of columns to string for easier use later
fight_df["weight_class"] = pd.Series(fight_df["weight_class"], dtype="string")
fight_df["winner"] = pd.Series(fight_df["winner"], dtype="string")
fight_df["event"] = pd.Series(fight_df["event"], dtype="string")
fight_df["method"] = pd.Series(fight_df["method"], dtype="string")
fight_df["round"] = pd.Series(fight_df["round"], dtype="string")

fight_df["red_Significant_strikes"] = pd.Series(fight_df["red_Significant_strikes"], dtype="string")
fight_df["blue_Significant_strikes"] = pd.Series(fight_df["blue_Significant_strikes"], dtype="string")
fight_df["red_Total_strikes"] = pd.Series(fight_df["red_Total_strikes"], dtype="string")
fight_df["blue_Total_strikes"] = pd.Series(fight_df["blue_Total_strikes"], dtype="string")


fight_df["red_Sig_Str_head"] = pd.Series(fight_df["red_Sig_Str_head"], dtype="string")
fight_df["red_Sig_Str_body"] = pd.Series(fight_df["red_Sig_Str_body"], dtype="string")
fight_df["red_Sig_Str_leg"] = pd.Series(fight_df["red_Sig_Str_leg"], dtype="string")

fight_df["blue_Sig_Str_head"] = pd.Series(fight_df["blue_Sig_Str_head"], dtype="string")
fight_df["blue_Sig_Str_body"] = pd.Series(fight_df["blue_Sig_Str_body"], dtype="string")
fight_df["blue_Sig_Str_leg"] = pd.Series(fight_df["blue_Sig_Str_leg"], dtype="string")

fight_df["red_Sig_Str_distance"] = pd.Series(fight_df["red_Sig_Str_distance"], dtype="string")
fight_df["red_Sig_Str_clinch"] = pd.Series(fight_df["red_Sig_Str_clinch"], dtype="string")
fight_df["red_Sig_Str_ground"] = pd.Series(fight_df["red_Sig_Str_ground"], dtype="string")

fight_df["blue_Sig_Str_distance"] = pd.Series(fight_df["blue_Sig_Str_distance"], dtype="string")
fight_df["blue_Sig_Str_clinch"] = pd.Series(fight_df["blue_Sig_Str_clinch"], dtype="string")
fight_df["blue_Sig_Str_ground"] = pd.Series(fight_df["blue_Sig_Str_ground"], dtype="string")
fight_df.info()
```

```python
# converting to int from string
fight_df["red_sig_stks_landed"] = fight_df["red_sig_stks_landed"].astype(int, errors = 'raise')
fight_df["red_sig_stks_attempted"] = fight_df["red_sig_stks_attempted"].astype(int, errors = 'raise')

fight_df["blue_sig_stks_landed"] = fight_df["blue_sig_stks_landed"].astype(int, errors = 'raise')
fight_df["blue_sig_stks_attempted"] = fight_df["blue_sig_stks_attempted"].astype(int, errors = 'raise')

fight_df["red_total_stks_landed"] = fight_df["red_total_stks_landed"].astype(int, errors = 'raise')
fight_df["red_total_stks_attempted"] = fight_df["red_total_stks_attempted"].astype(int, errors = 'raise')

fight_df["blue_total_stks_landed"] = fight_df["blue_total_stks_landed"].astype(int, errors = 'raise')
fight_df["blue_total_stks_attempted"] = fight_df["blue_total_stks_attempted"].astype(int, errors = 'raise')

fight_df["red_TDs_landed"] = fight_df["red_TDs_landed"].astype(int, errors = 'raise')
fight_df["red_TDs_attempted"] = fight_df["red_TDs_attempted"].astype(int, errors = 'raise')

fight_df["blue_TDs_landed"] = fight_df["blue_TDs_landed"].astype(int, errors = 'raise')
fight_df["blue_TDs_attempted"] = fight_df["blue_TDs_attempted"].astype(int, errors = 'raise')

fight_df["round"] = fight_df["round"].astype(int, errors = 'raise') 
fight_df.info()
```

```python
# creating percentage column for strike and takedown stats
fight_df["red_sig_stk_pct"] = (fight_df["red_sig_stks_landed"] / fight_df["red_sig_stks_attempted"]) 
fight_df["red_total_stk_pct"] = (fight_df["red_total_stks_landed"] / fight_df["red_total_stks_attempted"])
fight_df["red_TD_pct"] = (fight_df["red_TDs_landed"] / fight_df["red_TDs_attempted"])

fight_df["blue_sig_stk_pct"] = (fight_df["blue_sig_stks_landed"] / fight_df["blue_sig_stks_attempted"]) 
fight_df["blue_total_stk_pct"] = (fight_df["blue_total_stks_landed"] / fight_df["blue_total_stks_attempted"]) 
fight_df["blue_TD_pct"] = (fight_df["blue_TDs_landed"] / fight_df["blue_TDs_attempted"])
fight_df
```

```python
# Getting percentage for strikes landed by location and target 
fight_df["red_sig_stk_head_pct"] = (fight_df["red_sig_stks_head_landed"] / fight_df["red_sig_stks_head_attempted"])
fight_df["red_sig_stk_body_pct"] = (fight_df["red_sig_stks_body_landed"] / fight_df["red_sig_stks_body_attempted"])
fight_df["red_sig_stk_leg_pct"] = (fight_df["red_sig_stks_leg_landed"] / fight_df["red_sig_stks_leg_attempted"]) 

fight_df["blue_sig_stk_head_pct"] = (fight_df["blue_sig_stks_head_landed"] / fight_df["blue_sig_stks_head_attempted"])
fight_df["blue_sig_stk_body_pct"] = (fight_df["blue_sig_stks_body_landed"] / fight_df["blue_sig_stks_body_attempted"])
fight_df["blue_sig_stk_leg_pct"] = (fight_df["blue_sig_stks_leg_landed"] / fight_df["blue_sig_stks_leg_attempted"]) 

fight_df["red_sig_stk_distance_pct"] = (fight_df["red_sig_stks_distance_landed"] / fight_df["red_sig_stks_distance_attempted"])
fight_df["red_sig_stk_clinch_pct"] = (fight_df["red_sig_stks_clinch_landed"] / fight_df["red_sig_stks_clinch_attempted"]) 
fight_df["red_sig_stk_ground_pct"] = (fight_df["red_sig_stks_ground_landed"] / fight_df["red_sig_stks_ground_attempted"]) 

fight_df["blue_sig_stk_distance_pct"] = (fight_df["blue_sig_stks_distance_landed"] / fight_df["blue_sig_stks_distance_attempted"])
fight_df["blue_sig_stk_clinch_pct"] = (fight_df["blue_sig_stks_clinch_landed"] / fight_df["blue_sig_stks_clinch_attempted"])
fight_df["blue_sig_stk_ground_pct"] = (fight_df["blue_sig_stks_ground_landed"] / fight_df["blue_sig_stks_ground_attempted"])

fight_df
```

```python
# converting names to string
fight_df["red_name"] = pd.Series(fight_df["red_name"], dtype="string")
fight_df["blue_name"] = pd.Series(fight_df["blue_name"], dtype="string")
fight_df
```

```python
# dropping all the original columns that are no longer useful 
fight_df = fight_df.drop(columns=["red_Significant_strikes", "red_Total_strikes", "red_TDs", "red_Sig_Str_head", \
                                  "red_Sig_Str_body","red_Sig_Str_leg", "red_Sig_Str_distance", "red_Sig_Str_clinch", \
                                  "red_Sig_Str_ground", "blue_Significant_strikes", "blue_Total_strikes", "blue_TDs",\
                                  "blue_Sig_Str_head", "blue_Sig_Str_head", "blue_Sig_Str_body", "blue_Sig_Str_leg", \
                                  "blue_Sig_Str_distance", "blue_Sig_Str_clinch", "blue_Sig_Str_ground"])
fight_df
```

```python
# Begining to get fight time by converting the time "object" to string
fight_df["Time"] = pd.Series(fight_df["Time"], dtype="string")
```

```python
# Splitting into minutes and seconds
fight_df[["time_mins", "time_sec"]] = fight_df.Time.str.split(":",expand=True).astype(int)
fight_df
```

```python
# converting everything into one column in seconds
fight_df["Time_seconds"] = (fight_df["time_mins"] * 60) + fight_df["time_sec"]
fight_df
```

```python
# getting seconds of rounds completed before the final round 
fight_df["prior_round_seconds"] = (fight_df["round"] - 1) * 300
fight_df
```

```python
# for TBD fights, there will be negative 300 seconds in prior-rounds so this fixes for that
fight_df["prior_round_seconds"] = fight_df["prior_round_seconds"].replace(-300, 0)
fight_df
```

```python
# adding all prior round seconds and final round seconds together to get full fight time in seconds
fight_df["fight_seconds"] = fight_df["Time_seconds"] + fight_df["prior_round_seconds"]
fight_df
```

```python
# drop the other now-useless columns
fight_df = fight_df.drop(columns=["prior_round_seconds", "Time_seconds", "time_sec", "time_mins", "Time"])
fight_df
```

```python
# for control time, many rows contained "--" to signal null value, so replacing that with a usable format
fight_df["red_Control_time"] = fight_df["red_Control_time"].replace("--", "0:00")
fight_df["blue_Control_time"] = fight_df["blue_Control_time"].replace("--", "0:00")
fight_df
```

```python
# splitting up control time into minutes and seconds seconds column just like with fight time 
fight_df[["red_control_mins", "red_control_sec"]] = fight_df.red_Control_time.str.split(":",expand=True).astype(int)
fight_df[["blue_control_mins", "blue_control_sec"]] = fight_df.blue_Control_time.str.split(":",expand=True).astype(int)
fight_df
```

```python
# converting everything into one column in seconds
fight_df["red_control_seconds"] = (fight_df["red_control_mins"] * 60) + fight_df["red_control_sec"]
fight_df["blue_control_seconds"] = (fight_df["blue_control_mins"] * 60) + fight_df["blue_control_sec"]
fight_df
```

```python
# drop useless columns
fight_df = fight_df.drop(columns=["red_Control_time", "blue_Control_time", "red_control_mins", "red_control_sec", \
                                  "blue_control_mins", "blue_control_sec"])
fight_df = fight_df.fillna(0)
fight_df
```

```python
# for future steps, we will be summing up each fighters cumulative stats offensively and defensively....
fight_df["blue_KedD"] = fight_df["red_KDs"]
fight_df["blue_SubEsc"] = fight_df["red_Sub_Atts"]
fight_df["blue_reversed"] = fight_df["red_Reversals"]
fight_df["blue_TotStkAbs"] = fight_df["red_total_stks_landed"]
fight_df["blue_TotStkSeen"] = fight_df["red_total_stks_attempted"]
fight_df["blue_SigStkAbs"] = fight_df["red_sig_stks_landed"]
fight_df["blue_SigStkSeen"] = fight_df["red_sig_stks_attempted"]
fight_df["blue_TakenDown"] = fight_df["red_TDs_landed"]
fight_df["blue_TDsSeen"] = fight_df["red_TDs_attempted"]
fight_df["blue_SigHeadStkAbs"] = fight_df["red_sig_stks_head_landed"]
fight_df["blue_SigHeadStkSeen"] = fight_df["red_sig_stks_head_attempted"]
fight_df["blue_SigBodyStkAbs"] = fight_df["red_sig_stks_body_landed"]
fight_df["blue_SigBodyStkSeen"] = fight_df["red_sig_stks_body_attempted"]
fight_df["blue_SigLegStkAbs"] = fight_df["red_sig_stks_leg_landed"]
fight_df["blue_SigLegStkSeen"] = fight_df["red_sig_stks_leg_attempted"]
fight_df["blue_SigDistStkAbs"] = fight_df["red_sig_stks_distance_landed"]
fight_df["blue_SigDistStkSeen"] = fight_df["red_sig_stks_distance_attempted"]
fight_df["blue_SigClchStkAbs"] = fight_df["red_sig_stks_clinch_landed"]
fight_df["blue_SigClchStkSeen"] = fight_df["red_sig_stks_clinch_attempted"]
fight_df["blue_SigGndStkAbs"] = fight_df["red_sig_stks_ground_landed"]
fight_df["blue_SigGndStkSeen"] = fight_df["red_sig_stks_ground_attempted"]
fight_df["blue_CtrldSec"] = fight_df["red_control_seconds"]
fight_df["blue_FightSec"] = fight_df["fight_seconds"] 
fight_df["blue_RndEnd"] = fight_df["round"] 

# for each stat (such as knockdowns), we get the flip side for the fighter on the other end (knocked-down fighter)
fight_df["red_KedD"] = fight_df["blue_KDs"]
fight_df["red_SubEsc"] = fight_df["blue_Sub_Atts"]
fight_df["red_reversed"] = fight_df["blue_Reversals"]
fight_df["red_TotStkAbs"] = fight_df["blue_total_stks_landed"]
fight_df["red_TotStkSeen"] = fight_df["blue_total_stks_attempted"]
fight_df["red_SigStkAbs"] = fight_df["blue_sig_stks_landed"]
fight_df["red_SigStkSeen"] = fight_df["blue_sig_stks_attempted"]
fight_df["red_TakenDown"] = fight_df["blue_TDs_landed"]
fight_df["red_TDsSeen"] = fight_df["blue_TDs_attempted"]
fight_df["red_SigHeadStkAbs"] = fight_df["blue_sig_stks_head_landed"]
fight_df["red_SigHeadStkSeen"] = fight_df["blue_sig_stks_head_attempted"]
fight_df["red_SigBodyStkAbs"] = fight_df["blue_sig_stks_body_landed"]
fight_df["red_SigBodyStkSeen"] = fight_df["blue_sig_stks_body_attempted"]
fight_df["red_SigLegStkAbs"] = fight_df["blue_sig_stks_leg_landed"]
fight_df["red_SigLegStkSeen"] = fight_df["blue_sig_stks_leg_attempted"]
fight_df["red_SigDistStkAbs"] = fight_df["blue_sig_stks_distance_landed"]
fight_df["red_SigDistStkSeen"] = fight_df["blue_sig_stks_distance_attempted"]
fight_df["red_SigClchStkAbs"] = fight_df["blue_sig_stks_clinch_landed"]
fight_df["red_SigClchStkSeen"] = fight_df["blue_sig_stks_clinch_attempted"]
fight_df["red_SigGndStkAbs"] = fight_df["blue_sig_stks_ground_landed"]
fight_df["red_SigGndStkSeen"] = fight_df["blue_sig_stks_ground_attempted"]
fight_df["red_CtrldSec"] = fight_df["blue_control_seconds"]
fight_df["red_FightSec"] = fight_df["fight_seconds"] # fight seconds..
fight_df["red_RndEnd"] = fight_df["round"] # and the round ended will be the same for each fighter
```
