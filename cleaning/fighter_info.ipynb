```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Load in scraped fighter data
career_stats = pd.read_csv("../data/career_fighter_stats.csv")
career_stats
```

```python
# Dropping unused columns
career_stats = career_stats.drop(columns=["nickname", "Unnamed: 0"])
career_stats
```

```python
fighter_info = career_stats[["name", "DOB", "ufc_record", "height", "weight", "reach", "stance"]]
fighter_info
```

```python
no_height = fighter_info[fighter_info["height"].str.contains("-")]
no_height
```

```python
# Removing fighters that have no height listed
fighter_info = fighter_info[~fighter_info["height"].str.contains("-")]
fighter_info
```

```python
# Splitting feet and inches into seperate columns
fighter_info[["height_feet", "height_inches"]] = fighter_info.height.str.split("'", expand=True)
fighter_info
```

```python
# Stripping " from inches and converting height to int
fighter_info["height_inches"] = fighter_info["height_inches"].str.strip('"')
fighter_info["height_feet"] = fighter_info["height_feet"].astype(int)
fighter_info["height_inches"] = fighter_info["height_inches"].astype(int)
fighter_info
```

```python
# Calculating total height from feet and inches columns
fighter_info["height_in"] = (fighter_info["height_feet"] * 12) + fighter_info["height_inches"]
fighter_info
```

```python
# dropping old height columns
fighter_info = fighter_info.drop(columns=["height", "height_feet", "height_inches"])
fighter_info
```

```python
# Stripping " from reach
fighter_info["reach"] = fighter_info["reach"].str.strip('"')
fighter_info
```

```python
# splitting fighters into two seperate dataframes based on if they have reach listed...
no_reach = fighter_info[fighter_info["reach"].str.contains("-")]
no_reach
```

```python
# ... as a majorit of fighters have no reach listed, it was necessary to aproximate... 
reach = fighter_info[~fighter_info["reach"].str.contains("-")]
reach
```

```python
# ... by using height as a proxy for reach  
no_reach["reach"] = no_reach["height_in"]
no_reach
```

```python
# And then putting the dataframes back together with aproximated reach for fighters with none listed
frames = [reach, no_reach]
fighter_df = pd.concat(frames)
fighter_df
```

```python
fighter_df.info()
```

```python
# setting reach as int
fighter_df["reach"] = fighter_df["reach"].astype(int)
fighter_df.info()
```

```python
fighter_df["stance"].value_counts()
```

```python
fighter_df["stance"] = fighter_df["stance"].astype(str)
```

```python
# Seeing as the vast majority of fighters are orthadox, assume that nan values are orthadox...
fighter_df["stance"] = fighter_df["stance"].replace({"nan" : "Orthodox"})
fighter_df["stance"] = fighter_df["stance"].replace({" Orthodox" : "Orthodox"})
fighter_df["stance"] = fighter_df["stance"].replace({" Southpaw" : "Southpaw"})
fighter_df["stance"] = fighter_df["stance"].replace({" Switch" : "Switch"})
fighter_df["stance"] = fighter_df["stance"].replace({" Open Stance" : "Switch"}) # With open as switch
fighter_df["stance"] = fighter_df["stance"].replace({" Sideways" : "Switch"}) # and same with sideways
fighter_df["stance"].value_counts()
```

```python
# fighters with no date of birth makes it impossible to calculate age at time of fight later on...
no_dob = fighter_df[fighter_df["DOB"].str.contains("-")]
no_dob
```

```python
# ... so fighters with no DOB are dropped
clean_fighter_df = fighter_df[~fighter_df["DOB"].str.contains("-")]
```

```python
# Saving the clean dataframe
clean_fighter_df.to_csv("../data/clean_fighter_info.csv")
clean_fighter_df
```
