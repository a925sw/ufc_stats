```python
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
client = MongoClient()
```

```python
ufc_stats_db = client["ufc_stats_db"]
```

```python
past_fights = ufc_stats_db["past_fights"]
```

```python
big_df = pd.read_csv("../data/big_df.csv")
big_df = big_df.drop(columns="lev1")
big_df
```

```python
past_fights_df = big_df[big_df["winner"] != "tbd"]
past_fights_df
```

```python
past_fights.insert_many(past_fights_df.to_dict("records"))
```

```python
fighters_df = pd.read_csv("../data/fighters_df.csv")
fighters_df = fighters_df.drop(columns="Unnamed: 0")
fighters_df
```

```python
fighters = ufc_stats_db["fighters"]
```

```python
fighters.insert_many(fighters_df.to_dict("records"))
```
