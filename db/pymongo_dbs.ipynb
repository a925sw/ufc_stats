```python
import pandas as pd
import pymongo
from pymongo import MongoClient
import pprint
pd.set_option('display.max_columns', None)
```

```python
client = MongoClient()
```

```python
ufc_stats_db = client["ufc_stats_db"]
```

```python
past_fights = ufc_stats_db["past_fights"]
```

```python
big_df = pd.read_csv("../data/cumulative_df.csv")
big_df = big_df.drop(columns="Unnamed: 0")
big_df
```

```python
past_fights_df = big_df[big_df["red_winner"] != "tbd"]
past_fights_df
```

```python
past_fights_df["winner"] = past_fights_df["red_winner"]
past_fights_df = past_fights_df.drop(columns=["red_winner", "blue_winner"])
past_fights_df["weight_class"] = past_fights_df["red_weightclass"]
past_fights_df = past_fights_df.drop(columns=["red_weightclass", "blue_weightclass"])
past_fights_df["title_fight"] = past_fights_df["red_titleFight"]
past_fights_df = past_fights_df.drop(columns=["red_titleFight", "blue_titleFight"])
past_fights_df["event"] = past_fights_df["red_event"]
past_fights_df = past_fights_df.drop(columns=["red_event", "blue_event"])
past_fights_df["date"] = past_fights_df["red_date"]
past_fights_df = past_fights_df.drop(columns=["red_date", "blue_date"])
past_fights_df["method"] = past_fights_df["red_method"]
past_fights_df = past_fights_df.drop(columns=["red_method", "blue_method"])
past_fights_df["location"] = past_fights_df["red_location"]
past_fights_df = past_fights_df.drop(columns=["red_location", "blue_location"])
past_fights_df = past_fights_df.drop(columns=["red_weight", "blue_weight", "red_index", "blue_index"])
past_fights_df
```

```python
past_fights_df["method"].value_counts()
```

```python
past_fights.drop()
past_fights.insert_many(past_fights_df.to_dict("records"))
```
