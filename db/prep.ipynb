```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
cumulative_df = pd.read_csv("../data/cumulative_df.csv")
cumulative_df = cumulative_df.drop(columns="Unnamed: 0")
cumulative_df.head(20)
```

```python
cumulative_df = cumulative_df[cumulative_df["red_winner"] != "tbd"]
cumulative_df
```

```python
cumulative_df = cumulative_df.drop(columns=["red_event", "blue_event", "red_date", "blue_date", "red_winner", "blue_winner"])
```

```python
# creating copy df sorted by index
df1 = cumulative_df.copy()
df1 = df1.sort_index()
df1.head(20)
```

```python
# Splitting columns on hyphen ....
df1.columns = df1.columns.str.split('_', expand=True)
df1
```

```python
# in order to stack dataframe into levels on index and red/blue fighter
df1 = df1.stack(0).rename_axis(['lev1','lev2'])
df1# Now that each fight is stacked with nested levels for red and blue fighters, it is possible to sum up the total
# stats of each fighter by name REGARDLESS of if the were in the red or blue corner. Without this stacking, trying
# to sum up the total for any given stat (e.g takedowns) would return two numbers, one for their total when 
# fighting out of the red corner, and another for them fighting out of the blue corner. This turned into a bit of
# a conceptual challenge for me, so shout out to stackoverflow user jezrael for helping me with this.

# utilizing groupby["name"], i was able to get the sum of relevant stats for each fighter through their UFC careers
# and subtract the data of the fight in that row to get a cumulative sum COMING INTO THE FIGHT, so the future model
# would only have information about the fighters that was available BEFORE the fight took place
df1["sumTotStksLnd"] = (df1.iloc[::-1].groupby(["name"])["TotStksLnd"].cumsum())
df1["sumTotStksAtt"] = (df1.iloc[::-1].groupby(["name"])["TotStksAtt"].cumsum())
df1["sumSigStksLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStksLnd"].cumsum())
df1["sumSigStksAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStksAtt"].cumsum())
df1["sumTDsLnd"] = (df1.iloc[::-1].groupby(["name"])["TDsLnd"].cumsum()) 
df1["sumTDsAtt"] = (df1.iloc[::-1].groupby(["name"])["TDsAtt"].cumsum()) 
df1["sumSubAtt"] = (df1.iloc[::-1].groupby(["name"])["SubAtts"].cumsum())
df1["sumCtrlSec"] = (df1.iloc[::-1].groupby(["name"])["CtrlSec"].cumsum())
df1["sumKDs"] = (df1.iloc[::-1].groupby(["name"])["KDs"].cumsum()) 
df1["sumRevs"] = (df1.iloc[::-1].groupby(["name"])["Reversals"].cumsum())
df1["sumSigBodyLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkBodyLnd"].cumsum())
df1["sumSigBodyAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkBodyAtt"].cumsum())
df1["sumSigHeadLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkHeadLnd"].cumsum())
df1["sumSigHeadAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkHeadAtt"].cumsum())
df1["sumSigLegLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkLegLnd"].cumsum())
df1["sumSigLegAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkLegAtt"].cumsum())
df1["sumSigDistLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkDistLnd"].cumsum())
df1["sumSigDistAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkDistAtt"].cumsum())
df1["sumSigClchLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkClchLnd"].cumsum())
df1["sumSigClchAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkClchAtt"].cumsum())
df1["sumSigGndLnd"] = (df1.iloc[::-1].groupby(["name"])["SigStkGndLnd"].cumsum())
df1["sumSigGndAtt"] = (df1.iloc[::-1].groupby(["name"])["SigStkGndAtt"].cumsum())

df1["sumTotStksAbs"] = (df1.iloc[::-1].groupby(["name"])["TotStkAbs"].cumsum()) 
df1["sumTotStksSeen"] = (df1.iloc[::-1].groupby(["name"])["TotStkSeen"].cumsum())
df1["sumSigStksAbs"] = (df1.iloc[::-1].groupby(["name"])["SigStkAbs"].cumsum()) 
df1["sumSigStksSeen"] = (df1.iloc[::-1].groupby(["name"])["SigStkSeen"].cumsum())

df1["sumTDsAbs"] = (df1.iloc[::-1].groupby(["name"])["TakenDown"].cumsum())
df1["sumTDsSeen"] = (df1.iloc[::-1].groupby(["name"])["TDsSeen"].cumsum())

df1["sumSubEsc"] = (df1.iloc[::-1].groupby(["name"])["SubEsc"].cumsum()) 
df1["sumCtrldSec"] = (df1.iloc[::-1].groupby(["name"])["CtrldSec"].cumsum()) 

df1["sumKedD"] = (df1.iloc[::-1].groupby(["name"])["KedD"].cumsum()) 
df1["sumReversed"] = (df1.iloc[::-1].groupby(["name"])["reversed"].cumsum()) 

df1["sumSigBodyAbs"] = (df1.iloc[::-1].groupby(["name"])["SigBodyStkAbs"].cumsum()) 
df1["sumSigBodySeen"] = (df1.iloc[::-1].groupby(["name"])["SigBodyStkSeen"].cumsum())
df1["sumSigHeadAbs"] = (df1.iloc[::-1].groupby(["name"])["SigHeadStkAbs"].cumsum()) 
df1["sumSigHeadSeen"] = (df1.iloc[::-1].groupby(["name"])["SigHeadStkSeen"].cumsum())
df1["sumSigLegAbs"] = (df1.iloc[::-1].groupby(["name"])["SigLegStkAbs"].cumsum()) 
df1["sumSigLegSeen"] = (df1.iloc[::-1].groupby(["name"])["SigLegStkSeen"].cumsum()) 
df1["sumSigDistAbs"] = (df1.iloc[::-1].groupby(["name"])["SigDistStkAbs"].cumsum()) 
df1["sumSigDistSeen"] = (df1.iloc[::-1].groupby(["name"])["SigDistStkSeen"].cumsum())
df1["sumSigClchAbs"] = (df1.iloc[::-1].groupby(["name"])["SigClchStkAbs"].cumsum()) 
df1["sumSigClchSeen"] = (df1.iloc[::-1].groupby(["name"])["SigClchStkSeen"].cumsum())
df1["sumSigGndAbs"] = (df1.iloc[::-1].groupby(["name"])["SigGndStkAbs"].cumsum()) 
df1["sumSigGndSeen"] = (df1.iloc[::-1].groupby(["name"])["SigGndStkSeen"].cumsum())
df1["sumFightSec"] = (df1.iloc[::-1].groupby(["name"])["FightSec"].cumsum())

df1["sumDecWin"] = (df1.iloc[::-1].groupby(["name"])["decWin"].cumsum()) 
df1["sumSubWin"] = (df1.iloc[::-1].groupby(["name"])["subWin"].cumsum()) 
df1["sumKOWin"] = (df1.iloc[::-1].groupby(["name"])["KO/TKOWin"].cumsum())
df1["sumDecLoss"] = (df1.iloc[::-1].groupby(["name"])["DecLoss"].cumsum())
df1["sumSubLoss"] = (df1.iloc[::-1].groupby(["name"])["SubLoss"].cumsum())
df1["sumKOLoss"] = (df1.iloc[::-1].groupby(["name"])["KOLoss"].cumsum()) 
df1["wins"] = (df1.iloc[::-1].groupby(["name"])["win"].cumsum()) 
df1["losses"] = (df1.iloc[::-1].groupby(["name"])["loss"].cumsum()) 

df1["sumTitleFights"] = (df1.iloc[::-1].groupby(["name"])["titleFight"].cumsum()) 


df1
```

```python
df1["winPCT"] = df1["wins"] / (df1["wins"] + df1["losses"])
df1
```

```python
fighters_df = df1.groupby('name').first().reset_index()
fighters_df
```

```python
# calculating strikes landed per minute and accuracy over entire UFC career for red fighter
fighters_df["TotStkPerMin"] = fighters_df["sumTotStksLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["SigStkPerMin"] = fighters_df["sumSigStksLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["TotStkAcc"] = fighters_df["sumTotStksLnd"] / fighters_df["sumTotStksAtt"]
fighters_df["SigStkAcc"] = fighters_df["sumSigStksLnd"] / fighters_df["sumSigStksAtt"]

# red strikes landed per minute by target
fighters_df["StkHeadPerMin"] = fighters_df["sumSigHeadLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkBodyPerMin"] = fighters_df["sumSigBodyLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkLegPerMin"] = fighters_df["sumSigLegLnd"] / (fighters_df["sumFightSec"] / 60)

# red strikes landed per minute by range
fighters_df["StkDistPerMin"] = fighters_df["sumSigDistLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkClchPerMin"] = fighters_df["sumSigClchLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkGndPerMin"] = fighters_df["sumSigGndLnd"] / (fighters_df["sumFightSec"] / 60)

# red strikes absorbed per minute 
fighters_df["TotStkAbsPerMin"] = fighters_df["sumTotStksAbs"] / (fighters_df["sumFightSec"] / 60)
fighters_df["SigStkAbsPerMin"] = fighters_df["sumSigStksAbs"] / (fighters_df["sumFightSec"] / 60)

# red strikes absorbed per minute by target
fighters_df["StkHeadAbsPerMin"] = fighters_df["sumSigHeadAbs"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkBodyAbsPerMin"] = fighters_df["sumSigBodyAbs"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkLegAbsPerMin"] = fighters_df["sumSigLegAbs"] / (fighters_df["sumFightSec"] / 60)

# red strikes absorbed per minute by range
fighters_df["StkDistAbsPerMin"] = fighters_df["sumSigDistAbs"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkClchAbsPerMin"] = fighters_df["sumSigClchAbs"] / (fighters_df["sumFightSec"] / 60)
fighters_df["StkGndAbsPerMin"] = fighters_df["sumSigGndAbs"] / (fighters_df["sumFightSec"] / 60)

# red strikes landed differential
fighters_df["TotStkPerMinDiff"] = fighters_df["TotStkPerMin"] - fighters_df["TotStkAbsPerMin"]
fighters_df["SigStkPerMinDiff"] = fighters_df["SigStkPerMin"] - fighters_df["SigStkAbsPerMin"]

# red control and controled time in seconds per fight minute, and differential
fighters_df["CtrlSecPerMin"] = fighters_df["sumCtrlSec"] / (fighters_df["sumFightSec"] / 60)
fighters_df["CtrldSecPerMin"] = fighters_df["sumCtrldSec"] / (fighters_df["sumFightSec"] / 60)
fighters_df["CtrlDiff"] = fighters_df["CtrlSecPerMin"] - fighters_df["CtrldSecPerMin"]

# red Takedown accuracy and defense 
fighters_df["TDAcc"] = fighters_df["sumTDsLnd"] / fighters_df["sumTDsAtt"]
fighters_df["TDDef"] = 100 - (fighters_df["sumTDsAbs"] / fighters_df["sumTDsSeen"])

# red takedowns per minute landed and absorbed 
fighters_df["TDsPerMin"] = fighters_df["sumTDsLnd"] / (fighters_df["sumFightSec"] / 60)
fighters_df["TDsAbsPerMin"] = fighters_df["sumTDsAbs"] / (fighters_df["sumFightSec"] / 60)

# red knockdowns per 100 significant strikes landed
fighters_df["KDsPer100Stk"] = fighters_df["sumKDs"] / (fighters_df["sumSigStksLnd"] / 100)

fighters_df
```

```python
fighters_df = fighters_df.drop(columns=["CtrlSec", "CtrldSec", "DecLoss", "FightSec", "KDs", "KO/TKOWin", "KOLoss", "KedD",\
                                       "Reversals", "RndEnd", "SigBodyStkAbs", "SigBodyStkSeen", "SigClchStkAbs", \
                                        "SigClchStkSeen", "SigDistStkAbs", "SigDistStkSeen", "SigGndStkAbs", "SigGndStkSeen",\
                                       "SigHeadStkAbs", "SigHeadStkSeen", "SigLegStkAbs", "SigLegStkSeen", "SigStkAbs", \
                                       "SigStkBodyAtt", "SigStkBodyLnd", "SigStkBodyPct", "SigStkClchAtt", "SigStkClchLnd",\
                                       "SigStkClchPct", "SigStkDistAtt", "SigStkDistLnd", "SigStkDistPct", "SigStkGndAtt",\
                                       "SigStkGndLnd", "SigStkGndPct", "SigStkHeadAtt", "SigStkHeadLnd", "SigStkHeadPct",\
                                       "SigStkLegAtt", "SigStkLegLnd", "SigStkLegPct", "SigStkPct", "SigStkSeen", \
                                        "SigStksAtt", "SigStksLnd", "SubAtts", "SubEsc", "SubLoss", "TDPct", "TDsAtt",\
                                       "TDsLnd", "TDsSeen", "TakenDown", "TotStkAbs", "TotStkSeen", "TotStksAtt", \
                                        "TotStksLnd", "TotstkPct", "decWin", "reversed", "subWin", "titleFight", "win",\
                                        "loss"
                                       ])
fighters_df
```

```python
fighters_df.to_csv("../data/fighters_df.csv")
```
